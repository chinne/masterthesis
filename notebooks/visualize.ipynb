```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_raw = pd.read_csv('../data/raw/creditcard.csv')
```

```python
display(df_raw.shape, df_raw.columns, df_raw.head(),df_raw.describe())
```

```python
f,ax=plt.subplots(10,3,figsize=(12,60))
#f.delaxes(ax)
col = list(df_raw)
col = [e for e in col if e not in ('Class')]

for i,feature in enumerate(col):
    sns.distplot(df_raw[df_raw['Class']==1].dropna()[(feature)], ax=ax[i//3,i%3], kde_kws={"color":"black"}, hist=False, label = 'fraud' )
    sns.distplot(df_raw[df_raw['Class']==0].dropna()[(feature)], ax=ax[i//3,i%3], kde_kws={"color":"black"}, hist=False, label = 'normal' )

    # Get the two lines from the ax[i//2,i%2]es to generate shading
    l1 = ax[i//3,i%3].lines[0]
    l2 = ax[i//3,i%3].lines[1]

    # Get the xy data from the lines so that we can shade
    x1 = l1.get_xydata()[:,0]
    y1 = l1.get_xydata()[:,1]
    x2 = l2.get_xydata()[:,0]
    y2 = l2.get_xydata()[:,1]
    ax[i//3,i%3].fill_between(x2,y2, color="red", alpha=0.8)
    ax[i//3,i%3].fill_between(x1,y1, color="blue", alpha=0.8)

    #grid
    ax[i//3,i%3].grid(b=True, which='major', color='grey', linewidth=0.3)
    
    ax[i//3,i%3].set_title('{}'.format(feature), fontsize=18)
    ax[i//3,i%3].set_ylabel("fraction", fontsize=12)
    ax[i//3,i%3].set_xlabel(str(feature), fontsize=12)


plt.tight_layout()
plt.savefig('../reports/figures/creditcard_data_distribution.png')
plt.show()
```
